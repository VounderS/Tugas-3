```python
list = []
while True:
  print("""1.Menambahkan
2.Menghapus
3.Mengedit
4.Menampilkan
5.Mencari item
6.Mencari index
0.Menghentikan program""")

  a = int(input("Pilih menu :"))

  if a == 1:
    while True:
      list.append(input("Tambahkan item :"))
      b = input("Mau lanjut atau berhenti Y/N").lower()
      if b == "n":
        break 

  elif a == 2:
    while True:
      list.pop(int(input("Hapus item pada index ke :")))
      b = input("Mau lanjut atau berhenti Y/N").lower()
      print(list)
      if b == "n":
        break 

  elif a == 3:
    for x in range(len(list)):
      c = input("Masukkan nama item :")
      if list[x] == c:
        list[x] = input("Masukkan nama item yang akann mengganti :")
        print(list)
        break

      else:
        print("Item tidak ditemukan pada list")
        break

  elif a == 4:
    print(list)

  elif a == 5:
    d = input("Masukkan nama item yang dicari :")
    if d in list:
      print("Item ditemukan")
    else:
      print("item tidak ditemukan")

  elif a == 6:
    e = int(input("Masukkan index pada list :"))
    if list[e] in list:
      print("index ditemukan")
    else:
      print("index tidak ditemukan")

  else:
    break
```
